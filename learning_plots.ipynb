```python
# imports
import pandas as pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pandas.read_csv('data/training_data.csv', sep="\t")
```

```python
df
```

```python
df["voice"]=df["accent"] + "_" + df["gender"]
```

```python
sns.lineplot(data=df, x="Step", y="TrainingLoss", hue="voice")
```

```python
sns.lineplot(data=df, x="Step", y="ValidationLoss", hue="voice")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.lineplot(data=df, x="Step", y="TrainingLoss", hue="voice",ax=ax[0])
sns.lineplot(data=df, x="Step", y="ValidationLoss", hue="voice", ax=ax[1])

# change to have same y axis
min_val = min(df["TrainingLoss"].min(), df["ValidationLoss"].min())
max_val = max(df["TrainingLoss"].max(), df["ValidationLoss"].max())
min_val = min_val - 0.002
max_val = max_val + 0.002
ax[0].set_title("Training Loss")
ax[1].set_title("Validation Loss")
ax[0].set_ylim(min_val, max_val)
ax[1].set_ylim(min_val, max_val)
# remove legend from first plot
ax[0].legend().remove()
plt.show()
```
