```python
# imports
import pandas as pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pandas.read_csv('data/training_data.csv', sep="\t")
```

```python
df
```

```python
df["voice"]=df["accent"] + "_" + df["gender"]
```

```python
sns.lineplot(data=df, x="Step", y="TrainingLoss", hue="voice")
```

```python
sns.lineplot(data=df, x="Step", y="ValidationLoss", hue="voice")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.lineplot(data=df, x="Step", y="TrainingLoss", hue="voice",ax=ax[0])
sns.lineplot(data=df, x="Step", y="ValidationLoss", hue="voice", ax=ax[1])

# change to have same y axis
ax[0].set_ylim(0.35, 0.46)
ax[1].set_ylim(0.35, 0.46)
plt.show()
```
