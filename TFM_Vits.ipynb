## Vits, so far outperforms Finetuning of F5

```python
# Imports
from transformers import VitsModel, VitsTokenizer
import torch
from IPython.display import Audio
```

```python
# Argentinian


model = VitsModel.from_pretrained("ylacombe/mms-spa-finetuned-argentinian-monospeaker")
tokenizer = VitsTokenizer.from_pretrained("facebook/mms-tts-spa")
```

```python
#Chilean
model = VitsModel.from_pretrained("ylacombe/mms-spa-finetuned-chilean-monospeaker")
# tokenizer = VitsTokenizer.from_pretrained("facebook/mms-tts-spa")
```

```python
# Colombian

model = VitsModel.from_pretrained("ylacombe/mms-spa-finetuned-colombian-monospeaker")
```

```python
text_example = ('Hola soy una inteligencia artificial entrenada para hablar con un dialecto especifico del español.')
```

```python
inputs = tokenizer(text_example, return_tensors="pt")
input_ids = inputs["input_ids"]
```

```python
# no_grad() is required, otherwise throws this error:
# Can't call numpy() on Tensor that requires grad. Use tensor.detach().numpy() instead.
with torch.no_grad():
    outputs = model(input_ids)

```

```python
speech = outputs["waveform"]

Audio(speech, rate=16000)
```
